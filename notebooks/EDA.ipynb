```python
data["total score"]= data["math score"] + data["writing score"] + data["writing score"]
data["avg score"]= data["total score"]/3
data.head()
```

```python
plt.subplots(1,3,figsize=(25,6))
plt.subplot(141)
ax =sns.histplot(data=data,x='avg score',kde=True,hue='race/ethnicity')
plt.subplot(142)
ax =sns.histplot(data=data[data.gender=='female'],x='avg score',kde=True,hue='race/ethnicity')
plt.subplot(143)
ax =sns.histplot(data=data[data.gender=='male'],x='avg score',kde=True,hue='race/ethnicity')
plt.show()
```
